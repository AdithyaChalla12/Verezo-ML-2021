#### Performing classification using K-Nearest Neighbour

```python
import pandas as pd      #Pandas-Used to load the datafile as a Pandas dataframe and analyze it
```

```python
df = pd.read_csv(r"C:\Users\adith\OneDrive\Desktop\Adithya\My Projects\Linear Reg\Iris.csv")
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns='Id')
df
```

```python
#Prepare the training set
# X=feature values,all the columns except last column
x = df.iloc[:,:-1].values
# Y= target values,last column of the dT frame
y = df.iloc[:,4].values
x
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train_x)

train_x = scaler.transform(train_x)
test_x = scaler.transform(test_x)
train_x
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=5,metric='euclidean')
classifier.fit(train_x,train_y)
```

```python
pred_y = classifier.predict(test_x)
pred_y
```

```python
test_y
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_y, pred_y)
cm
```

```python
score = classifier.score(test_x, test_y)
score
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(test_y, pred_y))
print(classification_report(test_y, pred_y))
```

# Using SVM(Support Vector Machine)

```python
import pandas as pd 
```

```python
df = pd.read_csv(r"C:\Users\adith\OneDrive\Desktop\Adithya\My Projects\Linear Reg\Iris.csv")
df
```

```python
df = df.drop(columns='Id')
df
```

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
x
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(train_x,train_y)
```

```python
pred_y=classifier.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(confusion_matrix(test_y, pred_y))
print(accuracy_score(test_y, pred_y))
```
